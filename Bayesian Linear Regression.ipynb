```python
from grid import Grid
from regressor import Regressor
import numpy as np
import numpy.random
import numpy.ma
%pylab inline
```

```python
def phi_vector(grid, point):
    result = np.zeros(grid.get_size())
    for i, (l,idx) in enumerate(grid.get_sorted_keys()):
        result[i] = grid._basis(l,idx, point)
    return result
```

# 1d Parabla

```python
def parabola1d(x):
    return 4*x*(1-x)
size = 20 # number of data points
beta = 20 # precision
seed = 42
np.random.seed(seed)
x1 = np.random.uniform(0, 1, size)
x2 = np.random.uniform(0.25, 0.5, 3*size)
x = np.concatenate((x1,x2), axis=0)
x = np.sort(x)
data = x.reshape(-1, 1)
eps = np.random.randn(len(x))*np.sqrt(1.0/beta)
y = np.array(map(parabola1d,x))
t = y+eps
plot(x,t, '.')
```

```python
def plot_grid(level,dim, alpha):
    lmbda = alpha/beta
    grid = Grid(dim, level, 'lin')
    regressor = Regressor(grid, data, t, lmbda)
    Phi = regressor.create_B()
    dof = grid.get_size()
    
    SNInv = alpha*np.eye(dof) + beta*np.dot(Phi.T, Phi)
    mN = np.linalg.solve(SNInv, beta*np.dot(Phi.T, t))
    grid.set_alphas(mN)
    y = grid.dehierachization()
    
    xs = np.zeros([dof, 3])
    for i, (l,idx) in enumerate(grid.get_sorted_keys()):
        point = grid.point_to_coordinates(l,idx)
        xs[i,0] = point[0]
        phi = phi_vector(grid, point)
        sigma = np.linalg.solve(SNInv, phi)
        sigma = 1.0/beta+dot(phi.T, sigma)
        xs[i, 1] = y[(l,idx)]
        xs[i, 2] = np.sqrt(sigma)
        
    #prepare results
    idx_sort = np.argsort(xs[:,0])
    xs = xs[idx_sort, :]
    
    #plot
    plot(x,t, '.')
    errorbar(xs[:,0], xs[:,1], yerr=xs[:,2])
    fill_between(xs[:,0], xs[:,1]-xs[:,2], xs[:,1]+xs[:,2], interpolate=True, alpha=0.3, color='g')
    return xs


    
```

```python
level = 3
alpha = 0.001
dim = 1
plot(x,y, 'r-', hold=True)
xs = plot_grid(level, dim, alpha)
print xs
print "\nnote:The standrad deviation of %.3f is optimal"%(np.sqrt(1.0/beta))
```

# Online update

```python
def sort(x,y,t):
    idx = np.argsort(x)
    x = x[idx]
    y = y[idx]
    t = t[idx]
    return x,y,t

seed = 2000
beta = 20 # precision
level = 3
alpha = 15

np.random.seed(seed)
mb_size = 5

grid = Grid(dim, level, 'lin')
dof = grid.get_size()
S0Inv = alpha*np.eye(dof)
m0 = np.zeros(dof)

x = np.array([])
y = np.array([])
t = np.array([])


for it in xrange(10):
    x_mb = np.random.uniform(0, 1, mb_size)
    eps_mb = np.random.randn(len(x_mb))*np.sqrt(1.0/beta)
    y_mb = np.array(map(parabola1d,x_mb))
    t_mb = y_mb+eps_mb
    
    x = np.concatenate((x,x_mb), axis=0)
    y = np.concatenate((y,y_mb), axis=0)
    t = np.concatenate((t,t_mb), axis=0)
    
    x,y,t = sort(x,y,t)
    
    regressor = Regressor(grid, x_mb.reshape((-1,1)), t, 0)
    Phi = regressor.create_B()

    SNInv = S0Inv + beta*np.dot(Phi.T, Phi)
    mN = np.linalg.solve(SNInv, np.dot(S0Inv, m0) + beta*np.dot(Phi.T, t_mb))
    grid.set_alphas(mN)
    Y = grid.dehierachization()

    xs = np.zeros([dof, 3])
    for i, (l,idx) in enumerate(grid.get_sorted_keys()):
        point = grid.point_to_coordinates(l,idx)
        xs[i,0] = point[0]
        phi = phi_vector(grid, point)
        sigma = np.linalg.solve(SNInv, phi)
        sigma = 1.0/beta+dot(phi.T, sigma)
        xs[i, 1] = Y[(l,idx)]
        xs[i, 2] = np.sqrt(sigma)
    
    #prepare results
    idx_sort = np.argsort(xs[:,0])
    xs = xs[idx_sort, :]
    
    res = np.zeros(t.shape)
    for i in xrange(len(t)): 
        res[i] = grid.eval_point(x.reshape((-1,1))[i,:])

    #plot
    rmse = np.linalg.norm(res-t)/np.sqrt(len(t))
    print "Iteration: %i, stdmin: %.4f, stdmax: %.4f, rmse: %.4f"%(it, np.min(xs[:,2]), np.max(xs[:,2]), rmse)
    ylim(0,1.5)
    xlim(0,1)
    plot(x,y, 'r-')
    plot(x,t, 'b.')
    errorbar(xs[:,0], xs[:,1], yerr=xs[:,2])
    fill_between(xs[:,0], xs[:,1]-xs[:,2], xs[:,1]+xs[:,2], interpolate=True, alpha=0.3, color='g')
    figure()
    
    m0 = mN
    S0Inv = SNInv
#print xs
```
