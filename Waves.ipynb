# Drawing Degenerated Vibrations

```python
def wave(A1, A2, m,n,x,y):
    t = 1.
    C = 1.
    k = m*np.pi
    l = n*np.pi
    omega = C**2*(k**2 + l**2)
    return A1*np.sin(k*x)*np.sin(l*y) + A2*np.sin(l*x)*np.sin(k*y)
```

```python
def draw_wave(A1, A2, m,n, c=1):
    X = np.linspace(0, 1, 50, endpoint=True)
    Y = np.linspace(0, 1, 50, endpoint=True)

    X,Y = np.meshgrid(X, Y)

    wave_vec = np.vectorize(lambda x,y: wave(A1, A2, m,n,x,y))

    #pl.figure()
    Z = c*wave_vec(X,Y)
    ax = pl.subplot(1,1, 1, projection='3d')
    #ax.plot_surface(X, Y, , )
    ax.plot_wireframe(X, Y, Z, color='k', rstride=3, cstride=3, alpha=.3)
    cset = ax.contour(X, Y, Z, zdir='y', offset=1., cmap=cm.coolwarm)
    cset = ax.contour(X, Y, Z, zdir='z', offset=np.min(Z), cmap=cm.coolwarm)
    cset = ax.contour(X, Y, Z, zdir='x', offset=0., cmap=cm.coolwarm)


    ax.set_xlabel('$x_1$')
    ax.set_xlim(0, 1)
    ax.set_ylabel('$x_2$')
    ax.set_ylim(0, 1.1)
    ax.set_title('%d*(%d(%d, %d)+%d(%d,%d))'%(c, A1,m,n,A2,n,m))
#ax.set_zlim(bottom=-0.05)
```

```python
draw_wave(1,1,1,1)
```
