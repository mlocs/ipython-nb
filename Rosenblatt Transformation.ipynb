# Rosenblatt Transformation 

This notebook shows how to perform Rosenblatt transformation on your data to make them uniformly distributed over the unit square.  

*Note: as Rosenblatt transformation suffers if the random variables are correlated we will decorrelate/orthogonalize the data by computing left singular vectors. Both steps can cause a **significant loss of information**, hence causion is advised.*

```python
from statsmodels.nonparametric.kernel_density import *
import numpy as np
import scipy as sp
import scipy.io.arff
import sklearn as sk
import pandas as pd
from pandas.tools.plotting import scatter_matrix
```

We will use [statsmodels](http://statsmodels.sourceforge.net/devel/generated/statsmodels.nonparametric.kernel_density.KDEMultivariate.html) to compute CDFs and [pandas](http://pandas.pydata.org/pandas-docs/stable/visualization.html) for visualization. 

First we need to read data. This step is rather obvious.

```python
# read a small portion 
(data, meta) = sp.io.arff.loadarff('/path/to/file/DR5_MGS_ugrizeC_train.60000.arff')
data = np.array(map(list, data))
size = 1000 # data.shape[0] # here we take only the first 1000 entries to be faster. 
dim = data.shape[1]-1 # dimensionality
dat = data[:size, :dim] # the last column contains labels
```

```python
dat_orig_df = pd.DataFrame(dat, columns=meta.names()[:dim])
scatter_matrix(dat_orig_df, diagonal='kde')
```

As we can see, the first 5 dimensions are strongly correlated and hence the results of the Rosenblatt transformation are unsutisfying:

```python
def rosenblatt(dat):
    """Performs the Rosenblatt transformation of the dataset and returns the new dataset"""
    dim = dat.shape[1]
    kdes = np.empty(dim, dtype=object)
    kdes[0] = KDEMultivariate(dat[:,0], var_type='c')
    for d in xrange(1,dim):
        kdes[d] = KDEMultivariateConditional(endog=dat[:,d], exog=dat[:,:d], 
                  dep_type='c', indep_type='c'*d, bw='normal_reference')
    trans_dat = map(lambda i: kdes[i].cdf(), xrange(dim))
    return np.array(trans_dat).T
```

```python
dat_orig_transf_df = pd.DataFrame(rosenblatt(dat), columns=meta.names()[:dim])
scatter_matrix(dat_orig_transf_df, diagonal='kde')
```

We use the fact that the singular vectors are orthogonal to remove the correlation in data.

```python
U,_,_ = sp.linalg.svd(dat, full_matrices=False) # I don't need the full sizexsize matrix
dat_orthog_df = pd.DataFrame(U, columns=meta.names()[:dim])
scatter_matrix(dat_orthog_df, diagonal='kde')
```

Obviously, the correlation is removed, but the data are not spread as nicely as we would like. The outliers would also prohibit simple scaling/shifting to the unit cube. Let's see the results of the Rosenblat transformation now.

```python
# visualize the transformed data
dat_orthog_transf_df = pd.DataFrame(rosenblatt(U), columns=meta.names()[:dim])
scatter_matrix(dat_orthog_transf_df, diagonal='kde')
```

As I mentioned at the beginning, this transformation would not always result superior results, as we may loose important information at every step. But this is the price we always have to pay for changing data.
